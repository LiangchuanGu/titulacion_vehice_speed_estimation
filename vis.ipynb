```python
import os
import numpy as np
import matplotlib.pyplot as plt
import ast
import cv2
import argparse
from joblib import load
from matplotlib import rcParams
rcParams['font.family'] = ['serif']
rcParams['font.serif'] = ['Times New Roman']
rcParams['font.size'] = 14
```

```python
videos = [1,2,5,6,7,9,11,12,13,14]
all_video_data = []
all_video_line_data = []

for videoind in videos:
    video = f"source_material/streetVideos/seattle{videoind}.mp4"
    data = load(f"results/velocities/iou/velocities{videoind}")
    spotVelocities = load(f"results/spot_velocities/iou/spot_velocities{videoind}")

    all_samples_concat = np.hstack(list(data))
    all_video_data.append(all_samples_concat)
    all_video_line_data.append(spotVelocities)
```

```python
indices = list(range(1,11))
videoDir = f"results/plots_tables/iou"

print("saving speed boxplots")
plt.clf()
fig, ax = plt.subplots()
_ = ax.boxplot(np.array(all_video_data, dtype=object))
_ = ax.set_xlabel("Video ID")
_ = ax.set_ylabel("Speed (km/h)")
plt.savefig(f"{videoDir}/speedBoxPlotsAll.jpg")
```

```python
indices = list(range(1,11))
videoDir = f"results/plots_tables/iou"
plt.subplots_adjust(bottom=0.3)
print("saving speed boxplots")
plt.clf()
fig, ax = plt.subplots()
fig.subplots_adjust(bottom=0.2)
_ = ax.boxplot(np.array(all_video_data, dtype=object), 0, "")
_ = ax.set_xlabel("Video ID")
_ = ax.set_ylabel("Speed (km/h)")
plt.savefig(f"{videoDir}/speedBoxPlotsAll.jpg")
```
